```sql
WITH total_emissions_all AS (
  SELECT SUM(CAST(REPLACE(`GHG emissions mtons CO2e`, ',', '') AS DOUBLE)) as grand_total
  FROM emissions_data
),
top_10_states AS (
  SELECT SUM(state_emissions) as top_10_emissions
  FROM (
    SELECT state_abbr,
           SUM(CAST(REPLACE(`GHG emissions mtons CO2e`, ',', '') AS DOUBLE)) as state_emissions
    FROM emissions_data
    GROUP BY state_abbr
    ORDER BY state_emissions DESC
    LIMIT 10
  )
)
SELECT 
  ROUND((top_10_emissions / grand_total) * 100, 2) as top_10_pct_of_total
FROM top_10_states
CROSS JOIN total_emissions_all
```
