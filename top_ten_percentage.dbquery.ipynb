```sql
WITH grand AS (
  SELECT SUM(ghg_emissions_mtons) AS grand_total 
  FROM emissions_data_clean
),
top10 AS (
  SELECT SUM(ghg_emissions_mtons) AS top_10_emissions
  FROM (
    SELECT state_abbr, SUM(ghg_emissions_mtons) AS ghg_emissions_mtons
    FROM emissions_data_clean 
    GROUP BY state_abbr
    ORDER BY ghg_emissions_mtons DESC 
    LIMIT 10
  )
)
SELECT ROUND((top_10_emissions / grand_total) * 100, 2) AS top_10_pct_of_total
FROM top10 CROSS JOIN grand
```
